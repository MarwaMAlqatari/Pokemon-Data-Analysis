## Making changes to data

```python
## adding column of total for stats
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
##df['Total'] = df.iloc[:,4:10].sum(axix=1)
df.head(10)
```

```python
## best pokemon depending on total
df.sort_values('Total', ascending=False).head(1)
```

## Filtering Data

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
```

```python
df.loc[df['Name'].str.contains('Mega')].head(10) #for not containing df.loc[~df['Name'].str.contains('Mega')]
```

```python
import re
#df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)].head(10)
df.loc[df['Name'].str.contains('^pi[a-z]*', flags = re.I, regex=True)].head(10)
```

## Aggregate Statistics (Groupby)

```python
df.groupby(['Type 1']).mean().sort_values('Attack', ascending = False)
```

```python
df['count'] = 1
df.groupby(['Type 1']).count()['count']
```

```python
df_10 = df.drop(['#', 'Generation','Total','count','Legendary'], axis=1).head(10)
df_10.plot.bar(stacked=True,figsize=(10,10),x="Name",title='Level of Qualities in each Pokemon');
#xlabel('Pokemon')
#ylabel('Level')
#title('Level of Qualities in each Pokemon')
```

```python
df_10.plot(subplots=True, figsize=(8, 8),x="Name",title='Level of Qualities in each Pokemon');
```
